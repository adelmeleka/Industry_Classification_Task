## **Cleaning Dataset**

```
### Helper Functions

noise_list = set(stopwords.words("english"))
# noise detection
def remove_noise(input_text):
    words = word_tokenize(input_text)
    noise_free_words = list()
    i = 0;
    for word in words:
        if word.lower() not in noise_list:
            noise_free_words.append(word)
        i += 1
    noise_free_text = " ".join(noise_free_words)
    return noise_free_text

def lemetize_words(input_text):
    words = word_tokenize(input_text)
    new_words = []
    lem = WordNetLemmatizer()
    for word in words:
        word = lem.lemmatize(word, "v")
        new_words.append(word)
    new_text = " ".join(new_words)
    return new_text

def remove_special_keywords(input_text):
  special_keyword_list = ["county", "build", "ayr", "soa", "x", "global", "brand", "london", "k",
                          "contract", "day", "days", "qa", "paris", "france", "experience",
                          "require", "months", "fix", "term", "work", "per", "week", "hour",
                          "multinational", "company", "salary", "spanish", "opportunity",
                          "british", "firm", "visa", "available", "citizen", "start", "immediate", 
                          "eh", "amsterdam", "netherlands", "travel", "cayman", "islands", "tax", 
                          "free", "month", "hr", "kingston", "western", "europe", "oxford", "pr",
                          "b", "w", "saas", "deep", "executive", "abu", "dhabi", "dubai", "kuwait", 
                          "september", "uk", "full", "time", "india", "th", "october"]
  
   words = word_tokenize(input_text)
    noise_free_words = list()
    i = 0;
    for word in words:
        if word.lower() not in special_keyword_list:
            noise_free_words.append(word)
        i += 1
    noise_free_text = " ".join(noise_free_words)
    return noise_free_text
  
#### Main Function 
## cleaning 'job_title' records in a new column 
##in data frame:'cleaned_job_title'  

def dataset_cleaning(dataset):
    corpus = []
    
    for line in dataset["job_title"]:
        
        #transform all record into lower case
        review = re.sub('[^a-zA-Z]', ' ', line)
        review = review.lower()
         
        # remove non segnificant words
        #i.e: am, is, are ..ect
        review = remove_noise(review)
        
        #Lexicon Normalization: Lematization
        review = lemetize_words(review)
        
        #remove special keywords -> to do later
        review = remove_special_keywords(review)
        
        corpus.append(review)
    dataset["cleaned_job_title"] = corpus
    return
  
    
```

## **Linear SVM Classifier**

```
### Hyperparameter Tunning
## more params to tune better the model

def svc_param_selection(X, y, nfolds):
    Cs = [0.001, 0.01, 0.1, 1, 10]
    gammas = [0.001, 0.01, 0.1, 1]
#     degrees = [0, 1, 2, 3, 4, 5, 6]
    
    param_grid = {
                  'C': Cs,
                  'gamma' : gammas
                 }
    
    grid_search = GridSearchCV(svm.SVC(kernel='linear'), param_grid, cv=nfolds)
    
    grid_search.fit(X, y)
    
    grid_search.best_params_
    return grid_search.best_params_

  
### Testing 

best_params = svc_param_selection(x_train,y_train,7)

print(best_params)
```
